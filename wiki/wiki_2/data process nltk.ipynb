```python

from pathlib import Path
import torch
torch.cuda.empty_cache()
paths = [str(x) for x in Path('C:/langchain2/wiki/wiki_2/trext_data').glob('**/*.txt')]
paths
```

```python
# initialize an empty string to hold all the text
text = ""

# loop over the file paths and read each file into the text string
for path in paths:
    with open(path, 'r', encoding='utf-8') as f:
        text += f.read()
text = text.replace('\n', ' ')
```

```python
# Convert the text to lowercase
text = text.lower()
```

```python
text
```

```python
print(len(text))
```

```python
# Split the text into words
words = text.split()

# Remove single and double character words
filtered_words = [word for word in words if len(word) > 2]

# Join the remaining words back into a single string
processed_text = " ".join(filtered_words)
```

```python
print(len(processed_text))
```

```python
save_path = 'C:/langchain2/wiki/wiki_2/trext_data/processed_text.txt'
with open(save_path, 'w', encoding='utf-8') as f:
    f.write(processed_text)
```

```python
import re
# Define the regular expression to match the pattern of "[number]"
pattern = r'\[\d+\]'

# Use re.sub() to replace all instances of the pattern with an empty string
processed_text_1 = re.sub(pattern, '', processed_text)
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import SnowballStemmer

# Download necessary NLTK resources
nltk.download('stopwords')
nltk.download('punkt')

# Initialize stemmer and stop words
stemmer = SnowballStemmer('english')
stop_words = set(stopwords.words('english'))

# Example text data
text = text_1

# Tokenize the text into individual words
words = nltk.word_tokenize(text)

# Remove stop words and stem the remaining words
cleaned_words = []
for word in words:
    if word not in stop_words:
        cleaned_words.append(stemmer.stem(word))

# Convert the cleaned words back into a string
cleaned_text = ' '.join(cleaned_words)
print(cleaned_text)
```

```python
save_path = 'C:/langchain2/wiki/wiki_2/trext_data/processed_text.txt'
with open(save_path, 'w', encoding='utf-8') as f:
    f.write(cleaned_text)
```
