```python
import pandas as pd
```

```python
df = pd.read_csv('C:/langchain2/data_medical/generic_drug_processd.csv')
df_1 = pd.read_csv('C:/langchain2/data_medical/generic_drug_processd_1.csv')
```

```python
df = df.rename(columns={'Drug': 'intput_text'})
df = df.rename(columns={'Condition': 'output_text'})
df_1 = df_1.rename(columns={'Condition': 'intput_text'})
df_1 = df_1.rename(columns={'Drug': 'output_text'})
df = pd.concat([df, df_1], axis=0, ignore_index=True)
# Shuffle the rows of the combined DataFrame
df = df.sample(frac=1).reset_index(drop=True)
```

```python
df = df.astype(str)
```
