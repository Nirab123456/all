```python
import pandas as pd 
```

```python
drug_df = pd.read_csv('C:/langchain2/data_medical/drugand_med\Drug_clean.csv')
```

```python
drug_df
```

```python
collum_names= ['EaseOfUse','Effective','Form','Indication','Price','Reviews','Satisfaction','Type']
```

```python
drug_df = drug_df.drop(collum_names, axis=1)
```

```python
drug_df
```

```python
# Assuming you have a DataFrame called 'drug_df' and a specific column name 'column_name'
specific_string = "which medication is used for "
drug_df['Condition'] = specific_string + drug_df['Condition'].astype(str)
# lower case
drug_df = drug_df.applymap(lambda x: x.lower() if isinstance(x, str) else x)
```

```python
drug_df
```

```python
drug_df = drug_df.astype(str)
```

```python
drug_df
```

```python
new_column_order = ['Drug', 'Condition']
```

```python
# Reorder the columns in the DataFrame
drug_df = drug_df[new_column_order]
```

```python
drug_df
```

```python
drug_df.to_csv('C:/langchain2/data_medical/generic_drug_processd_1.csv', index=False)
```
